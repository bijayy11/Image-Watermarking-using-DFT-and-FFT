resizing image into square matrix which is also a power of 2

```python
def optimum_power_of_2(num):
    n=np.ceil(np.log2(num))
    N=2**n
    m=np.sqrt(N)
    m1=int(np.ceil(np.log2(m)))
    M1=2**m1
    return M1
```

```python
M1=optimum_power_of_2(gray_imageArray.size)
```

```python
def padding_With_Zeros(num,M1):
    twoPowerArray=np.zeros([M1,M1],dtype=float)
    size=int(np.sqrt(gray_imageArray.size))

    for i in range(size):
        for j in range(size):
            twoPowerArray[i,j]=gray_imageArray[i,j]
    return twoPowerArray     
```
